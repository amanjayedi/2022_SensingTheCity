### Generate Lat-Long Information

```python
## Make a featurelayer so the data can be worked with in memory.
PointsFL = os.path.join(env.workspace,'PointsFL')
arcpy.MakeFeatureLayer_management (PointsPath, PointsFL)
```

To create the lat-long info for the points, you have add fields to the feature layer and then calculate the geometry to populate the information

Add new fields: Long, Lat, Index to the 'LocationPoints' featurelayer. For this project, you do not need the Index function. However, it helps when we have multiple zones to analyze. Here, we'll keep the index as 0

```python
## First, check if the field to be added already exists.
fieldList = arcpy.ListFields(PointsFL)
fieldExists = False
for field in fieldList:
    if field.name == (("Long", "Lat", "Index")):
        fieldExists = True
        print ('Field already exists...no need to add it again')
        
## if the field doesn't exist...
if fieldExists == False:
    ## Add a new field to the featurelayer.
    arcpy.AddField_management(PointsFL, "Long", "FLOAT")
    arcpy.AddField_management(PointsFL, "Lat", "FLOAT")
    arcpy.AddField_management(PointsFL, "Index", "FLOAT")
```

Then use calculate geometry attributes to populate the information. The code can be found here, # Calculate Geometry Attributes of Lat and Long # https://pro.arcgis.com/en/pro-app/latest/tool-reference/data-management/calculate-geometry-attributes.htm

```python
#Calculagte geometry attributes
#arcpy.management.CalculateGeometryAttributes(in_features, geometry_property, {length_unit}, {area_unit}, {coordinate_system}, {coordinate_format})
arcpy.management.CalculateGeometryAttributes(in_features=PointsFL, geometry_property =[["Long", "POINT_X"], ["Lat", "POINT_Y"]])
```
