```python
#https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
classifier = pipeline('sentiment-analysis', model = model_name)
```

```python
SentimentResults = []
# for list in CaptionData[0:10]:
#   results = classifier([list])
#   for result in results:
#     print(result)

for list in CaptionData:
  results = classifier([list])
  for result in results:
    print(result)
    SentimentResults.append(result)
print(SentimentResults)
```

```python
#making list of each variables

label= []
score= []

for i in SentimentResults[0:len(SentimentResults)]:
    label.append(str(i['label']))
    score.append(str(i['score']))
print(label,score)
```

```python
PositiveCount = label.count('POSITIVE')
print(PositiveCount)
NegativeCount = label.count('NEGATIVE')
print(NegativeCount)
```

```python
#Convert the list into panda dataframe and write a backup csv file
import pandas as pd
import os
workspace = '/content/drive/MyDrive/MIZZOU/CAADRIA2022/Paper2/SentimentAnalysis/SentimentOutputResults'
csv_file_ex = os.path.join(workspace,"1_Sentiment results_Positive_Negative.csv")
list_dict = {'Sentiment': label, 'Score': score}
df = pd.DataFrame(list_dict)
df
try: 
    df.to_csv(csv_file_ex, index=False,encoding='utf-8')
    print('Data is written in the folder')
except IOError:
    print('File is not written in the folder')
```
